```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
file_paths = {
    "weekly_forecast": "weekly pre-dispatch forecast.csv",
    "train_data": "train_dataframes.xlsx",
    "test_data": "test_dataframes.xlsx",
    "continuous_data": "continuous dataset.csv",
}

weekly_forecast_df = pd.read_csv(file_paths["weekly_forecast"])
continuous_data_df = pd.read_csv(file_paths["continuous_data"])
train_data_df = pd.read_excel(file_paths["train_data"])
test_data_df = pd.read_excel(file_paths["test_data"])
```

```python
train_data_df["datetime"] = pd.to_datetime(train_data_df["datetime"])
test_data_df["datetime"] = pd.to_datetime(test_data_df["datetime"])


features = ["week_X-2", "week_X-3", "week_X-4", "MA_X-4", "dayOfWeek", "hourOfDay", "T2M_toc"]
target = "DEMAND"

X_train = train_data_df[features]
y_train = train_data_df[target]

X_test = test_data_df[features]
y_test = test_data_df[target]
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(X_train_scaled, y_train)
```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_scaled, y_train)
```

```python
y_pred_knn = knn_model.predict(X_test_scaled)
y_pred_rf = rf_model.predict(X_test_scaled)
```

```python
y_pred_hybrid = (0.5 * y_pred_knn) + (0.5 * y_pred_rf)
```

```python
def evaluate_model(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    return {"Model": model_name, "MAE": mae, "RMSE": rmse, "R²": r2}
```

```python
knn_results = evaluate_model(y_test, y_pred_knn, "KNN")
rf_results = evaluate_model(y_test, y_pred_rf, "Random Forest")
hybrid_results = evaluate_model(y_test, y_pred_hybrid, "Hybrid Model")
```

```python
model_comparison = pd.DataFrame([knn_results, rf_results, hybrid_results])
print("Model Performance Comparison:")
print(model_comparison)
```

```python
plt.figure(figsize=(12, 5))
plt.plot(y_test.values, label="Actual Demand", color="black", linestyle="dashed")
plt.plot(y_pred_knn, label="KNN Predicted", color="blue", alpha=0.7)
plt.plot(y_pred_rf, label="RF Predicted", color="red", alpha=0.7)
plt.plot(y_pred_hybrid, label="Hybrid Predicted", color="green", alpha=0.7)
plt.legend()
plt.xlabel("Time Index")
plt.ylabel("Electricity Demand")
plt.title("Actual vs. Predicted Demand (KNN, RF, Hybrid)")
plt.show()
```

```python
knn_errors = y_test - y_pred_knn
rf_errors = y_test - y_pred_rf
hybrid_errors = y_test - y_pred_hybrid
```

```python
plt.figure(figsize=(12, 5))
plt.hist(knn_errors, bins=30, alpha=0.5, label="KNN Errors", color="blue")
plt.hist(rf_errors, bins=30, alpha=0.5, label="RF Errors", color="red")
plt.hist(hybrid_errors, bins=30, alpha=0.5, label="Hybrid Errors", color="green")
plt.axvline(0, color="black", linestyle="dashed", linewidth=1)
plt.legend()
plt.xlabel("Prediction Error")
plt.ylabel("Frequency")
plt.title("Error Distribution (KNN, RF, Hybrid)")
plt.show()
```

```python
predictions_df = pd.DataFrame({
    "datetime": test_data_df["datetime"],  
    "Actual Demand": y_test,
    "KNN Predicted": y_pred_knn,
    "Random Forest Predicted": y_pred_rf,
    "Hybrid Predicted": y_pred_hybrid
})
```
